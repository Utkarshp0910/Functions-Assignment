```python
#Ques 1- What is the difference between a function and a method in Python?"""
"""Answer - Function: A block of reusable code that works independently and can be called anywhere in your program.
Method: A function tied to an object. It is called on that object and can only work with its data."""

#Ques 2- Explain the concept of function arguments and parameters in Python.
"""Answer - In Python, parameters are variables listed in a function's definition to receive inputs, while arguments are the actual values passed to the function when you call it."""

#Ques 3- What are the different ways to define and call a function in Python?
"""Answer - Functions in Python are defined using the def keyword, followed by a name, parameters in parentheses, and a body containing the code to execute. They are called by writing the function name followed by parentheses, with any required arguments inside the parentheses."""

#Ques 4- What is the purpose of the `return` statement in a Python function?
"""Answer - The `return` statement in a Python function is used to send a value back to the caller, allowing the function to produce a result that can be stored or used in further operations."""

#Ques 5- What are iterators in Python and how do they differ from iterables?
"""Answer - Iterators in Python are objects that allow you to traverse through a sequence one element at a time using the `next()` function. Iterables are objects like lists, tuples, or strings that can be looped over, but they need to be converted to an iterator using the `iter()` function to access elements one by one."""

#Ques 6- Explain the concept of generators in Python and how they are defined.
"""Answer - Generators in Python are a type of iterable, created using functions with the `yield` statement. Unlike regular functions, which return a value once and stop, generators yield a value each time the `next()` function is called, allowing them to produce a series of values lazily, one at a time. They are defined by using the `def` keyword and `yield` within the function body."""

#Ques 7- What are the advantages of using generators over regular functions?
"""Answer - Generators offer several advantages over regular functions, including memory efficiency by generating values one at a time, reducing the need to store all values in memory at once. They also support lazy evaluation, producing values only when needed, which improves performance for large data sets. Additionally, generators are convenient for representing infinite sequences, as they can generate values on demand without causing memory overflow, unlike regular functions that would require storing all values."""

#Ques 8- What is a lambda function in Python and when is it typically used?
"""Answer - A lambda function in Python is a small, anonymous function defined using the `lambda` keyword, followed by parameters, a colon, and an expression. It is typically used for short, throwaway functions where defining a full function with `def` would be unnecessary, such as when passing functions as arguments to higher-order functions like `map()`, `filter()`, or `sorted()`."""

#Ques 9- Explain the purpose and usage of the `map()` function in Python
"""Answer - The `map()` function in Python is used to apply a given function to all items in an iterable (such as a list or tuple) and returns an iterator that produces the results. It is typically used when you want to transform or modify each element in an iterable without using an explicit loop. The syntax is `map(function, iterable)`, where `function` is the function to apply, and `iterable` is the collection of items."""

#Ques 10- What is the difference between map(), reduce(), and filter() functions in Python?
"""Answer - The difference between `map()`, `reduce()`, and `filter()` functions in Python is as follows:

- `map()`: Applies a given function to each item in an iterable and returns an iterator of the results. It is used for transforming data.
  Example: `map(function, iterable)`

- `reduce()`: Applies a given function cumulatively to the items in an iterable, reducing the iterable to a single value. It is typically used for aggregation or combining values.
  Example: `reduce(function, iterable)`

- `filter()`: Applies a given function to each item in an iterable and returns an iterator of the items that evaluate to `True` (i.e., the function returns `True`). It is used for filtering data.
  Example: `filter(function, iterable)`"""

#Ques 11-  Using pen & Paper write the internal mechanism for sum operation using  reduce function on this given list:[47,11,42,13];
"""Answer - https://github.com/Utkarshp0910/Answer-11 """

#Practical Questions

#Ques 1-  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#Answer -
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(result)  # Output will be 12 (2 + 4 + 6)

#Ques 2- Create a Python function that accepts a string and returns the reverse of that string.
#Answer -
def reverse_string(s):
    return s[::-1]
input_string = "hello"
result = reverse_string(input_string)
print(result)  # Output will be "olleh"

#Ques 3-  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
#Answer -
def square_numbers(numbers):
    return [num ** 2 for num in numbers]
numbers = [1, 2, 3, 4, 5]
result = square_numbers(numbers)
print(result)  # Output will be [1, 4, 9, 16, 25]

#Ques 4- Write a Python function that checks if a given number is prime or not from 1 to 200.
#Answer -
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
for number in range(1, 201):
    if is_prime(number):
        print(number)

#Ques 5-  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#Answer -
class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # First two terms of Fibonacci sequence
        self.count = 0  # To keep track of the number of terms generated

    def __iter__(self):
        return self  # Return the iterator itself

    def __next__(self):
        if self.count < self.n:
            fib = self.a
            self.a, self.b = self.b, self.a + self.b  # Update the next two Fibonacci numbers
            self.count += 1
            return fib
        else:
            raise StopIteration  # Stop when the specified number of terms is reached
fibonacci = FibonacciIterator(10)  # Create an iterator for the first 10 Fibonacci numbers
for num in fibonacci:
    print(num)

#Ques 6 - Write a generator function in Python that yields the powers of 2 up to a given exponent.
#Answer -
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_two(5):
    print(power)

#Ques 7 - Implement a generator function that reads a file line by line and yields each line as a string.
#Answer - def read_file_line_by_line(filename):
with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
filename = 'example.txt'
for line in read_file_line_by_line(filename):
    print(line)

#Ques 8 - Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
#Answer -
# List of tuples
tuples = [(1, 4), (3, 1), (5, 2), (4, 3)]

# Sort the list of tuples based on the second element of each tuple using lambda
sorted_tuples = sorted(tuples, key=lambda x: x[1])

# Print the sorted list
print(sorted_tuples)
[(3, 1), (5, 2), (4, 3), (1, 4)]

#Ques 9 - Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
#Answer -
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Use map() to convert the Celsius temperatures to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the result
print(fahrenheit_temperatures)

#Ques 10 -
#Answer -
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    # Use filter to remove characters that are vowels
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)

#Ques 11 - """Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this: Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.
#Answer - def calculate_order_value(orders):
    result = []
    for order in orders:
        order_number, price_per_item, quantity = order
        total_value = price_per_item * quantity
        if total_value < 100:
            total_value += 10  # Add 10€ if total value is less than 100
        result.append((order_number, total_value))
    return result

# Example usage
orders = [
    [1, 10, 5],    # Order 1: 10€ per item, 5 items
    [2, 20, 3],    # Order 2: 20€ per item, 3 items
    [3, 50, 2],    # Order 3: 50€ per item, 2 items
    [4, 80, 1]     # Order 4: 80€ per item, 1 item
]

result = calculate_order_value(orders)
print(result)



```
